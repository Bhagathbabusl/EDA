```python
import pandas as pd

# Define the file path
file_path = r"application_train.csv"

# Read the CSV file into a DataFrame
application_train = pd.read_csv(file_path)

# Print the shape of the DataFrame
print("Shape of the CSV file:", application_train.shape)
```

```python
import pandas as pd

# Define the file path
file_path = r"HomeCredit_columns_description.xlsx"

# Read the CSV file into a DataFrame
HomeCredit_columns_description = pd.read_excel(file_path)

# Print the shape of the DataFrame
print("Shape of the CSV file:", HomeCredit_columns_description.shape)
```

```python
# Separate numerical and categorical columns
numerical_columns = application_train.select_dtypes(include=['int64', 'float64'])
categorical_columns = application_train.select_dtypes(exclude=['int64', 'float64'])

# Print the lists of numerical and categorical columns
print("Numerical Columns:")
print(numerical_columns.columns.tolist())

print("\nCategorical Columns:")
print(categorical_columns.columns.tolist())
```

```python
application_train_cat=application_train[categorical_columns.columns.tolist()].head()
```

```python
application_train_cat
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for i in application_train_cat.columns:
    encoded_labels = label_encoder.fit_transform(application_train_cat[i])
    application_train_cat[i]=encoded_labels
```
